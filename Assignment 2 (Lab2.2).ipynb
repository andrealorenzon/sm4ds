```r
n <- 50
K <- 4
M <- 6
y <- matrix(0, M, n )
# generate the values
for (m in 1:M){
  y[m, ] <- sample(1:K, n, replace=TRUE, prob = c( 7/16, 5/16, 3/16, 1/16) )
} 
observed_matrix <- apply( y,1, table)
chisq.test(observed_matrix, p = c( 7/16, 5/16, 3/16, 1/16) )
```

### A better challanger appears...

```r
#add a row for another player with better chances
y<-rbind(y,sample(1:K, n, replace=TRUE, prob = c( 5/16, 3/16, 5/16, 3/16) ))
```

```r
#recompute the test
observed_matrix <- apply( y,1, table)
chisq.test(observed_matrix, p = c( 7/16, 5/16, 3/16, 1/16) )
```

When the good challanger appears, $\chi^2$ value explodes, and p-value drops to significance level.

```r
#for verification, plot the density of the shots for all players
require(ggplot2)
require(reshape2)
shots <- data.frame(t(y))
shots=melt(shots,id.vars=NULL)
ggplot(shots,aes(x=value, fill=variable))+geom_density(alpha=.1)
```
